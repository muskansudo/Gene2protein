```python
%%R
install.packages("BiocManager")
BiocManager::install("ensembldb")
BiocManager::install("AnnotationHub")
```

```python
%%R
library(ensembldb)
library(AnnotationHub)

ah <- AnnotationHub()
edb <- ah[["AH119325"]]
```

```python
%%R
gene_id <- "ENSG00000106868"

tx <- transcripts(edb, filter = ~ gene_id == gene_id)
tx_ids <- tx$tx_id

tx_ids
```

```python
%%R
tx <- transcripts(edb,
    filter = ~ gene_id == "ENSG00000106868",
    columns = c("tx_id", "tx_biotype", "tx_cds_seq_start",
                "tx_cds_seq_end", "gene_id", "tx_support_level",
                "tx_id_version", "gc_content", "tx_external_name",
                "tx_is_canonical", "tx_name")
)

tx_df <- as.data.frame(tx)

write.table(tx_df,
            "BGLAP_transcripts.tsv",
            sep="\t", row.names=FALSE, quote=FALSE)

tx_df
```

```python
%%R
prot <- proteins(edb, filter = ~ gene_id == "ENSG00000106868")
prot_ids <- prot$protein_id

prot_ids
```

```python
%%R
tx_ranges <- transcripts(edb,
    filter = ~ gene_id == "ENSG00000106868",
    columns = c("tx_id", "tx_biotype", "tx_cds_seq_start",
                "tx_cds_seq_end", "gene_id", "tx_support_level",
                "tx_id_version", "gc_content", "tx_external_name",
                "tx_is_canonical", "tx_name")
)

tx_ranges
```

```python
%%R
prot <- proteins(edb,
    filter = ~ gene_id == "ENSG00000106868",
    columns = c("tx_id", "protein_id", "protein_sequence", "gene_name")
)

prot_df <- as.data.frame(prot)

write.table(prot_df,
            "BGLAP_proteins.tsv",
            sep="\t", row.names=FALSE, quote=FALSE)

prot_df
```

```python
%%R
mapping_df <- as.data.frame(tx_ranges)[, c("seqnames", "start", "end", "width", "strand", "tx_id")]

mapping_df
```

```python
%%R
write.table(mapping_df,
            "BGLAP_mapping.tsv",
            sep="\t", row.names=FALSE, quote=FALSE)
```
